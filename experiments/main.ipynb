```python
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# inicializujeme knižnicu seaborn
sns.set()
```

```python
#ako dlho trval ezxperiment
from datetime import datetime

print(data["timestamp"].min())
print(data["timestamp"].max())
print("Hours:", (data["timestamp"].max() - data["timestamp"].min()) / (60 * 60))
print("Days:", (data["timestamp"].max() - data["timestamp"].min()) / (60 * 60 * 24))

print(datetime.utcfromtimestamp(data["timestamp"].min()).strftime('%Y-%m-%d %H:%M:%S'))

print(datetime.utcfromtimestamp(data["timestamp"].max()).strftime('%Y-%m-%d %H:%M:%S'))


```

```python
#ide dlho idk co s tym nepouzivaj zatial
for i in data["timestamp"]:
    data["timestamp_date"] = datetime.utcfromtimestamp(i).strftime('%Y-%m-%d %H:%M:%S')
print(data)
```

```python
#normalizacia timestamp
from sklearn import preprocessing as sp

scaler = sp.MinMaxScaler(feature_range=(0, 12))
scaler.fit(np.array(data["timestamp"]).reshape(-1, 1))
data["timestamp_days"] = scaler.transform(np.array(data["timestamp"]).reshape(-1, 1))
display(data)
```

```python
#preskalovanie delty na hodiny
data["delta_days"] = (data["delta"] / (60 * 60 * 24))
```

```python
#pocet zucastnenych studentov
print((data["user_id"].unique()).size)
print((data["user_id"].unique()))
```

```python
#pocet zapisov kazdeho studenta
users = pd.value_counts(data["user_id"])
users
```

```python
#studentov co maju menej nez 100 zapisov zmazeme
data = data.groupby('user_id').filter(lambda x: len(x) > 100)
```
